```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/logistic/emnist-letters-train.csv', header= None)
train = df_train.to_numpy()
df_test = pd.read_csv('/content/drive/MyDrive/logistic/emnist-letters-test.csv', header= None)
test = df_test.to_numpy()
```

```python
input1 = train[0:60000, 1:]/255 #Scaling values between 0 and 1
ylabels = train[0:60000, 0]
```

```python
x0 = np.ones([np.shape(input1)[0], 1])
input = np.concatenate((x0, input1),axis = 1).T
print(np.shape(input))
```

```python
uniquech = set(ylabels)
logisticY = np.zeros([np.size(ylabels), len(uniquech)])
for i in range(np.size(ylabels)):
  for j in range(1, len(uniquech)+1):
    if ylabels[i] == j:
      logisticY[i,j-1] = 1
logisticY = logisticY.T
print(logisticY)
```

```python
def sigmoid(z):
  return 1/ (1+np.exp(-z))
```

```python
def train(x, y, alpha, iterations):

#initialise thetas
  w1 = np.random.random([x.shape[1], 50])*0.01
  w2 = np.random.random([50, len(uniquech)])*0.01
  b1 = np.ones([1, 50], dtype = int)
  b2 = np.ones([1, len(uniquech)], dtype = int)
  for iter in range(iterations):

#forward
    z2 = np.dot(x, w1) + b1
    a2 = sigmoid(z2)
    z3 = np.dot(a2, w2) + b2
    out = sigmoid(z3)
    
#errors
    d3 = out - y
    d2 = (np.dot(d3 ,w2.T)) * (a2*(1 - a2))

#backward
    pd_w2 = np.dot(a2.T, d3) / (x.shape[0])
    pd_w1 = np.dot(x.T , d2) / (x.shape[0])
    pd_b2 = np.sum(d3) / (x.shape[0])
    pd_b1 = np.sum(d2) / (x.shape[0])
    w1 = w1 - (alpha*pd_w1) 
    w2 = w2 - (alpha*pd_w2) 
    b1 = b1 - (alpha*pd_b1)
    b2 = b2 - (alpha*pd_b2)


  return w1, w2, out
```

```python
w1, w2, out = train(x=input, y=logisticY, alpha = 0.01, iterations=200)
```

```python
def accuracy(label, preds):
  truthtable = (label == preds).astype(int)
  truthlist = truthtable.tolist()
  accuracy = (truthlist.count([1]))*(100/(np.shape(preds)[0]))
  print("Accuracy is", accuracy , "%" )
  return accuracy
```
